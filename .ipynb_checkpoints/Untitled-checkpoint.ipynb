```python
# OpenCV, NumPy, Matplotlib のインポート
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/Users/shipro/Documents/CREATE_DATA2/Options/20230528_co_second_q02_no_46_Options.png")

# 1. グレースケール変換
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 2. 二値化 (大津の手法)
thresh_val, bin_img = cv2.threshold(
    gray,
    0,
    255,
    cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU
)

# 3. モルフォロジー演算 (ノイズ除去等)
kernel = np.ones((3,3), np.uint8)
bin_img = cv2.morphologyEx(bin_img, cv2.MORPH_OPEN, kernel, iterations=1)
# bin_img = cv2.morphologyEx(bin_img, cv2.MORPH_CLOSE, kernel, iterations=1)
# 必要に応じてパラメータや種類を調整

# 4. 輪郭抽出
contours, hierarchy = cv2.findContours(
    bin_img,
    cv2.RETR_EXTERNAL,
    cv2.CHAIN_APPROX_SIMPLE
)

# 5. 輪郭を元画像に描画 (オーバーレイ)
drawn_img = img.copy()
cv2.drawContours(drawn_img, contours, -1, (0, 0, 255), 2)  # 赤色で描画

# 6. 輪郭線だけの画像を作成
#    真っ黒な背景画像を用意し、そこに白色や緑色などで輪郭を描画
contour_img = np.zeros_like(img)  # 元画像と同じサイズで、RGB(3ch)の黒画像
cv2.drawContours(contour_img, contours, -1, (0, 255, 0), 1)  # 緑色で細く描画

# Matplotlib で表示 (OpenCV: BGR → Matplotlib: RGB に注意)
plt.figure(figsize=(12, 8))

# (1) グレースケール画像
plt.subplot(2, 2, 1)
plt.title("Gray")
plt.imshow(gray, cmap='gray')
plt.axis('off')

# (2) 二値化後
plt.subplot(2, 2, 2)
plt.title("Binary (Otsu)")
plt.imshow(bin_img, cmap='gray')
plt.axis('off')

# (3) 輪郭をオーバーレイした画像
plt.subplot(2, 2, 3)
plt.title("Contours Overlay")
drawn_img_rgb = cv2.cvtColor(drawn_img, cv2.COLOR_BGR2RGB)
plt.imshow(drawn_img_rgb)
plt.axis('off')

# (4) 輪郭線のみの画像
plt.subplot(2, 2, 4)
plt.title("Contours Only")
contour_img_rgb = cv2.cvtColor(contour_img, cv2.COLOR_BGR2RGB)
plt.imshow(contour_img_rgb)
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 画像読み込み
img = cv2.imread("/Users/shipro/Documents/CREATE_DATA2/Options/20230528_co_second_q02_no_46_Options.png")

# 1. グレースケール変換
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 2. 二値化 (大津の手法)
thresh_val, bin_img = cv2.threshold(
    gray,
    0,
    255,
    cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU
)

# 3. モルフォロジー演算 (ノイズ除去等)
kernel = np.ones((3,3), np.uint8)
bin_img = cv2.morphologyEx(bin_img, cv2.MORPH_OPEN, kernel, iterations=1)
# bin_img = cv2.morphologyEx(bin_img, cv2.MORPH_CLOSE, kernel, iterations=1)
# 必要に応じてパラメータや種類を調整

# 4. 輪郭抽出
contours, hierarchy = cv2.findContours(
    bin_img,
    cv2.RETR_EXTERNAL,
    cv2.CHAIN_APPROX_SIMPLE
)

# 5. 輪郭を元画像に描画 (オーバーレイ)
drawn_img = img.copy()
cv2.drawContours(drawn_img, contours, -1, (0, 0, 255), 2)  # 赤色で描画

# 6. 輪郭線だけの画像を作成 (Contours Only)
contour_img = np.zeros_like(img)  # 元画像と同じサイズの黒画像
cv2.drawContours(contour_img, contours, -1, (0, 255, 0), 1)  # 緑色で細く描画

# 7. 複数輪郭の最小/最大XY座標を求める (バウンディングボックス)
height, width = bin_img.shape[:2]
min_x, min_y = width, height
max_x, max_y = 0, 0

for cnt in contours:
    x, y, w, h = cv2.boundingRect(cnt)
    if x < min_x:
        min_x = x
    if y < min_y:
        min_y = y
    if (x + w) > max_x:
        max_x = x + w
    if (y + h) > max_y:
        max_y = y + h

# バウンディングボックスの値を表示
print(f"Bounding box: min_x={min_x}, min_y={min_y}, max_x={max_x}, max_y={max_y}")

# 8. バウンディングボックスを drawn_img に追加 (Contours + BBox)
drawn_img_with_bbox = drawn_img.copy()
cv2.rectangle(
    drawn_img_with_bbox,
    (min_x, min_y),
    (max_x, max_y),
    (255, 0, 255),  # マゼンタ
    2
)

# 9. バウンディングボックスのみを描画 (BoundingBox Only)
#    輪郭線は不要で、背景黒の画像にbboxだけを描画する
bbox_only_img = np.zeros_like(img)
cv2.rectangle(
    bbox_only_img,
    (min_x, min_y),
    (max_x, max_y),
    (255, 255, 0),  # 水色
    2
)

# Matplotlib で表示 (OpenCV: BGR → Matplotlib: RGB に注意)
plt.figure(figsize=(16, 10))

# (1) グレースケール画像
plt.subplot(2, 3, 1)
plt.title("Gray")
plt.imshow(gray, cmap='gray')
plt.axis('off')

# (2) 二値化後
plt.subplot(2, 3, 2)
plt.title("Binary (Otsu)")
plt.imshow(bin_img, cmap='gray')
plt.axis('off')

# (3) 輪郭をオーバーレイした画像
plt.subplot(2, 3, 3)
plt.title("Contours Overlay")
drawn_img_rgb = cv2.cvtColor(drawn_img, cv2.COLOR_BGR2RGB)
plt.imshow(drawn_img_rgb)
plt.axis('off')

# (4) Contours Only
plt.subplot(2, 3, 4)
plt.title("Contours Only")
contour_img_rgb = cv2.cvtColor(contour_img, cv2.COLOR_BGR2RGB)
plt.imshow(contour_img_rgb)
plt.axis('off')

# (5) Contours + BoundingBox
plt.subplot(2, 3, 5)
plt.title("Contours + BoundingBox")
drawn_img_with_bbox_rgb = cv2.cvtColor(drawn_img_with_bbox, cv2.COLOR_BGR2RGB)
plt.imshow(drawn_img_with_bbox_rgb)
plt.axis('off')

# (6) BoundingBox Only
plt.subplot(2, 3, 6)
plt.title("BoundingBox Only")
bbox_only_img_rgb = cv2.cvtColor(bbox_only_img, cv2.COLOR_BGR2RGB)
plt.imshow(bbox_only_img_rgb)
plt.axis('off')

plt.tight_layout()
plt.show()
```
