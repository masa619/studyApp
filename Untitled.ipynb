```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def auto_crop_whitespace(bin_image: np.ndarray) -> (int, int, int, int):
    """
    二値画像 bin_image から上下左右の「すべてが同じ値(0または255)」な行列を裁ち落として、
    残った範囲の [top, bottom, left, right] を返す。
    端ギリギリに図形(白画素)があればその手前でカットが止まる。
    """
    # 入力画像の高さ, 幅
    h, w = bin_image.shape[:2]

    # 行方向に上から下へスキャン
    top = 0
    for row in range(h):
        row_data = bin_image[row, :]
        # row_dataが全て0(黒) もしくは全て255(白)なら余白とみなしカット
        if np.all(row_data == row_data[0]):
            top += 1
        else:
            break
    
    # 行方向に下から上へスキャン
    bottom = h - 1
    for row in range(h - 1, -1, -1):
        row_data = bin_image[row, :]
        if np.all(row_data == row_data[0]):
            bottom -= 1
        else:
            break

    # 列方向に左から右へスキャン
    left = 0
    for col in range(w):
        col_data = bin_image[:, col]
        if np.all(col_data == col_data[0]):
            left += 1
        else:
            break

    # 列方向に右から左へスキャン
    right = w - 1
    for col in range(w - 1, -1, -1):
        col_data = bin_image[:, col]
        if np.all(col_data == col_data[0]):
            right -= 1
        else:
            break

    # 安全チェック (top > bottom などにならないように)
    top = max(top, 0)
    bottom = min(bottom, h - 1)
    left = max(left, 0)
    right = min(right, w - 1)

    return top, bottom, left, right


#---------------------------
# メインの処理
#---------------------------
img = cv2.imread("/Users/shipro/Documents/CREATE_DATA2/Options/20231029_co_second_q01_no_43_Options.png")


# 1. グレースケール変換
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 2. 二値化(大津)
_, bin_img = cv2.threshold(
    gray, 0, 255,
    cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU
)

# 3. モルフォロジー演算 (必要に応じて調整)
kernel = np.ones((3,3), np.uint8)
bin_img = cv2.morphologyEx(bin_img, cv2.MORPH_OPEN, kernel, iterations=1)
# bin_img = cv2.morphologyEx(bin_img, cv2.MORPH_CLOSE, kernel, iterations=1)

# (A) 余白トリミング (auto_crop_whitespace)
top, bottom, left, right = auto_crop_whitespace(bin_img)

# トリミング後の二値画像・カラー画像を作成
cropped_bin = bin_img[top:bottom+1, left:right+1]
cropped_img = img[top:bottom+1, left:right+1]

# (B) 輪郭抽出
contours, _ = cv2.findContours(cropped_bin, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 輪郭描画(オーバーレイ用)
drawn_cropped = cropped_img.copy()
cv2.drawContours(drawn_cropped, contours, -1, (0,0,255), 2)

# バウンディングボックスの計算
minX, minY = cropped_img.shape[1], cropped_img.shape[0]
maxX, maxY = 0, 0

for cnt in contours:
    x, y, w, h = cv2.boundingRect(cnt)
    if x < minX:
        minX = x
    if y < minY:
        minY = y
    if (x + w) > maxX:
        maxX = x + w
    if (y + h) > maxY:
        maxY = y + h

print(f"[Cropped range] top={top}, bottom={bottom}, left={left}, right={right}")
print(f"[BoundingBox in cropped] minX={minX}, minY={minY}, maxX={maxX}, maxY={maxY}")

# (C) マージンを追加
margin = 10
mh, mw = cropped_img.shape[:2]

# マージン付きでクロップ（さらにクロップ）
x1 = max(0, minX - margin)
y1 = max(0, minY - margin)
x2 = min(mw, maxX + margin)
y2 = min(mh, maxY + margin)

print(f"[Margin adjusted] x1={x1}, y1={y1}, x2={x2}, y2={y2}")

final_img = cropped_img[y1:y2, x1:x2]

#---------------------------
# 可視化 (matplotlib)
#---------------------------
plt.figure(figsize=(14, 8))

# 1. 二値化後(全体)
plt.subplot(2, 3, 1)
plt.title("Binary (Otsu) - Full")
plt.imshow(bin_img, cmap='gray')
plt.axis('off')

# 2. Cropped (二値)
plt.subplot(2, 3, 2)
plt.title("Cropped Binary")
plt.imshow(cropped_bin, cmap='gray')
plt.axis('off')

# 3. Cropped + Contours
plt.subplot(2, 3, 3)
plt.title("Cropped + Contours")
dcrgb = cv2.cvtColor(drawn_cropped, cv2.COLOR_BGR2RGB)
plt.imshow(dcrgb)
plt.axis('off')

# 4. final_img (マージン込み最終トリミング画像)
plt.subplot(2, 3, 4)
plt.title("Final Cropped w/ Margin")
final_rgb = cv2.cvtColor(final_img, cv2.COLOR_BGR2RGB)
plt.imshow(final_rgb)
plt.axis('off')

# 5. bounding box描画だけ確認 (オーバーレイ用)
drawn_bbox = drawn_cropped.copy()
cv2.rectangle(drawn_bbox, (minX, minY), (maxX, maxY), (255, 0, 255), 2)
# 同時にマージン範囲も描画 (オプション)
cv2.rectangle(drawn_bbox, (x1, y1), (x2, y2), (255, 255, 0), 2)

plt.subplot(2, 3, 5)
plt.title("Bounding Box + Margin Box")
drawn_bbox_rgb = cv2.cvtColor(drawn_bbox, cv2.COLOR_BGR2RGB)
plt.imshow(drawn_bbox_rgb)
plt.axis('off')

plt.tight_layout()
plt.show()
```
